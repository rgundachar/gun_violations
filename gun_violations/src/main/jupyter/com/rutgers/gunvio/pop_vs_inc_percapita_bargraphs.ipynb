```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.plotly as py
import plotly.graph_objs as go
#read csv file and create DF for population
state_population_df = pd.read_csv("../../../../resources/data/statepopulation.csv")
state_population_df.head()
```

```python
#read csv file and create DF for gun owners
gun_owners_df = pd.read_csv("../../../../resources/data/gunowners.csv")
state_population_df.head()
```

```python
#merge two df's on state 
state_vs_gunowners_df = pd.merge(state_population_df, gun_owners_df, on='States', how='inner')
#state_vs_gunowners= pd.DataFrame({'calcversionofpercapita': (int(state_vs_gunowners['# of guns registered'])*1000)/int(state_vs_gunowners['Population'])})
state_vs_gunowners_df.head()
```

```python
#read incidents file and merge with the above to DF to get incidents count by state 
incidents_by_state_df = pd.read_csv("../../../../resources/data/inc_dt_reg_lat_lon.csv")
incidents_by_state_df = pd.DataFrame({'IncidentsCount': incidents_by_state_df.groupby(['state'])['incident_id'].size()})
incidents_by_state_df = incidents_by_state_df.reset_index()
incidents_by_state_df.rename(columns={'state': 'States'}, inplace=True)
state_vs_gunowners_vs_inc_df = pd.merge(state_vs_gunowners_df, incidents_by_state_df, on='States', how='inner')
state_vs_gunowners_vs_inc_df.head()
state_vs_gunowners_vs_inc_df.set_index('States', inplace=True)
state_vs_gunowners_vs_inc_df["Population"] = state_vs_gunowners_vs_inc_df["Population"].str.replace(",","").astype(float)
state_vs_gunowners_vs_inc_df["# of guns registered"] = state_vs_gunowners_vs_inc_df["# of guns registered"].str.replace(",","").astype(float)
#state_vs_gunowners_vs_inc_df= pd.DataFrame({'calcversionofpercapita': (state_vs_gunowners_vs_inc_df['# of guns registered'])*1000/state_vs_gunowners_vs_inc_df['Population']})
state_vs_gunowners_vs_inc_df["calcversionofpercapita"] = (state_vs_gunowners_vs_inc_df['# of guns registered'])*1000/state_vs_gunowners_vs_inc_df['Population']
state_vs_gunowners_vs_inc_df = state_vs_gunowners_vs_inc_df.reset_index()
state_vs_gunowners_vs_inc_df
```

```python
#plot the graph against pop vs inc vs guns registered
#ax = state_vs_gunowners_vs_inc_df[['Population','# of guns registered','IncidentsCount']].plot(kind='bar',stacked=True, title ="pop vs in vs gun registered", figsize=(20, 10), legend=True, fontsize=12)
#ax.set_xlabel("States", fontsize=12)
#ax.set_ylabel("Counts of Incidents", fontsize=12)
#plt.show()
x=state_vs_gunowners_vs_inc_df['calcversionofpercapita']*1000
y=state_vs_gunowners_vs_inc_df['IncidentsCount']
size=y*0.006
scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
trace0 = go.Scatter(
    x=x,
    y=y,
    text=state_vs_gunowners_vs_inc_df['States'],
    mode='markers',
    marker=dict(
        color=['rgb(255, 255, 255)'],
        size=size,
       )
)
layout = go.Layout(
    title='Incidents Count Vs Per Capita for last 5 yrs',
    xaxis=dict(
        title='Per Capita',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Incidents Counts',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)
data = [trace0]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig ,filename='bubblechart-color')
```
