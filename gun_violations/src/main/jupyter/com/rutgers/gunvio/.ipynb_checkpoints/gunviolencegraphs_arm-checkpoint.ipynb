```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.patches as mpatches
from pandas.tools.plotting import table
```

```python

# File to Load (Remember to change these)
filetoread = "inc_dt_region_vict.csv"
gender_age_filetoread = "gender_age.csv"


# Read the City and Ride Data
agegroup_df = pd.read_csv(gender_age_filetoread)




# Read the City and Ride Data
region_vict_df = pd.read_csv(filetoread)
agegroup_df.head()

region_vict_df.head()
```

```python
#Convert the data type of Date of Incident
region_vict_df['Date of Incident'] = pd.to_datetime(region_vict_df['date'], errors='coerce')
region_vict_df['Year'] = region_vict_df['Date of Incident'].dt.year
region_vict_df['Month'] = region_vict_df['Date of Incident'].dt.month
region_vict_df['Day'] = region_vict_df['Date of Incident'].dt.day
region_vict_df['Qtr'] = region_vict_df['Date of Incident'].dt.quarter
region_vict_df['casualty'] = region_vict_df['n_killed']+region_vict_df['n_injured']

#Convert the data type of Date of Incident
agegroup_df['Date of Incident'] = pd.to_datetime(agegroup_df['date'], errors='coerce')
agegroup_df['Year'] = agegroup_df['Date of Incident'].dt.year
agegroup_df['Month'] = agegroup_df['Date of Incident'].dt.month
agegroup_df['Day'] = agegroup_df['Date of Incident'].dt.day
agegroup_df['Qtr'] = agegroup_df['Date of Incident'].dt.quarter
agegroup_df['casualty'] = agegroup_df['n_killed']+agegroup_df['n_injured']

```

```python
agegroup_master_df=agegroup_df
agegroup_master_df.head()
```

```python
#summarizin and pivoting to gett he data to adataframe to show year wise the incidents associated with age groups
agegroup_grp=agegroup_df.groupby(["Year","age"],as_index=False)["incident_id"].count()
agegroup_grp=agegroup_grp.rename(columns={'incident_id':'incidents'})
agegroup_grp.head()
```

```python
#summarizin and pivoting to gett he data to adataframe to show year wise the incidents associated with age groups
ax_df=agegroup_grp.pivot_table(index='Year', columns='age',values=['incidents'])

ax_df.head()
```

```python
#plot to show year wise the incidents associated with age groups
ax_df.plot.bar(rot=1,subplots=False,Width=.25,edgecolor='black',stacked=True,colormap='Paired',position=0,legend='',figsize=(12,9),sort_columns=True)
plt.title("Total Incidents By Age Group Each year ")

plt.ylabel("Total Incidents")

#red_patch = mpatches.Patch(color='red', label='The red data')
#plt.legend(handles=[red_patch])
lgnd=plt.legend(loc='upper right',markerscale=.5,title="Incidents by Age Group")

lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
lgnd.legendHandles[3]._sizes = [30]
lgnd.legendHandles[4]._sizes = [30]
lgnd.legendHandles[5]._sizes = [30]
lgnd.legendHandles[6]._sizes = [30]
plt.grid()
plt.tight_layout()
plt.savefig("stacked for age group.png")
```

```python
# plotting to show the age group thais involved in  more incidents
#building frames

bins2 = [0, 9, 19, 29, 39, 49,59,69,100]
#work_pur_grp=work_pur_grp.rename(columns={"Purchase ID":"Purchase Count"})
#agegroup_df=agegroup_df.rename({'incident_id':'incidents'})
agegroup_inc_df = pd.read_csv(gender_age_filetoread,usecols =["age","incident_id"])
group_names = ["<10", "10-19", "20-29", "30-39", "40-49", "50-59", "60-69","70+"]
agegroup_inc_df["age"] = pd.cut(agegroup_inc_df["age"], bins2, labels=group_names)
agegroup_inc_df.head()
```

```python
#grping and renaming, calculating percentages, renaming columns
inc_count=agegroup_inc_df.count()
total_inc_count=inc_count[1]
print
agegroup_inc_grp_df=agegroup_inc_df.groupby(["age"],as_index=False)["incident_id"].count()
agegroup_inc_grp_df=agegroup_inc_grp_df.rename(columns={'incident_id':'Incident Counts'})
agegroup_inc_grp_df["Percent of total Incidents"]=round((agegroup_inc_grp_df['Incident Counts']/total_inc_count)*100,1)
agegroup_inc_grp_df.head(100)
```

```python
# plotting to show the age group thais involved in  more incidents

fig = plt.figure()
fig.set_figheight(9)
fig.set_figwidth(18)
ax = fig.add_subplot(121)



type_slice=[age for age in agegroup_inc_grp_df["age"]]
per_slice=[per for per in agegroup_inc_grp_df["Percent of total Incidents"]]
colors = ["pink","orange","red","cyan","green","blue","brown","teal","lightblue"]
explode = (0,0,0.1,0,0,0,0,0)
plt.figsize=(14,12)
plt.pie(per_slice, explode=explode, labels=type_slice, colors=colors,autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Incidents By Age Group")

ax = fig.add_subplot(122)
plt.axis('off')
tbl = table(ax, agegroup_inc_grp_df, loc='center')
tbl.auto_set_font_size(True)
tbl.set_fontsize(18)
plt.title("% of Incidents By Age Group")

plt.savefig("Percentage of incidents by age groups.png")
plt.show()


```
